<h6> Converting the Column Names to lower Case and displaying the Null Values in each field

```python
data.columns = data.columns.str.lower() #converting columns names to lower type
print(data.isnull().sum()) # Printing the count of Null Values against values
sns.heatmap(data.isnull()); ## Visual Representation on getting the columns with Null Values
```

## 1- Missing Value Handling

Handling Null Values - 

<b> 1 - Eng V </b>

```python
newdf = data.copy()
engv_df_N = data[pd.isnull(data.engv)]# --- NaN Dataframe
engv_df_NN = data[~pd.isnull(data.engv)]# --- NotNull Dataframe
#print(engv_df_N.describe(include='all')) # --  Finding the describe method
#print(engv_df_NN.describe(include='all'))
#print(engv_df_N[engv_df_N['car']=='Nissan'].groupby(by=['model','body','engtype','drive'])['car'].count()) # Checking on Nissan Values
#engv_df_NN[engv_df_NN['car']=='Nissan'].groupby(by=['model','body','engtype'])['car'].count()
engv_df_N['engtype'].value_counts()
```

```python
#Replacing the Null Values of drive with median
newdf['engv'].fillna(newdf['engv'].median(),inplace=True, axis=0)
newdf.isnull().sum()
```

<b> 2 - Drive </b>

```python
#Replacing the Null Values of drive with Mode of the same
newdf['drive'].fillna(newdf['drive'].mode()[0],inplace=True) # Mode is "Full"
newdf.isnull().sum()
```

## 2 - Handling Inconsistent, Duplicates and Outliers

```python
newdf.describe(include='all')
```

<h6> 1 -  Dropping Duplicates Values </h6> 

```python
 #Dropping Duplicates
newdf.drop_duplicates(keep='first', inplace=True)
```

**2- Price** <br>
It has Zero Values which is very strange and will not help in any of the analysis. So we need to relook on the same

```python
pricezerodf = newdf[newdf['price'] == 0] # - To check count of Zero Price Value
print(pricezerodf.shape)
pricezerodf.describe(include='all')
```

**3-Mileage** <br>
It has Zero Values which is very strange and will not help in any of the analysis. So we need to relook on the same

```python
## Dropping the zero price as it is spread across all the models and we cannot predict the price for the same
newdf.drop(labels=newdf[newdf['price'] == 0].index,inplace=True)
```

```python
# 2 - Zero Values in Mileage
mileagezerodf = newdf[newdf['mileage'] == 0]
print(mileagezerodf.shape)
```

```python
## Dropping the zero mileage as it is spread across all the models
newdf.drop(labels=newdf[newdf['mileage'] == 0].index,inplace=True)
```

```python
 #Dropping Duplicates
newdf.drop_duplicates(keep='first', inplace=True)
```

<h6>4 - Year </h6> is in different format needs to be changed to Date Format

```python
#Changing the Year to Datetype
newdf['year'] = pd.to_datetime(newdf['year'], format='%Y')
```

```python
def CarCat(cat):
    if cat == 'BMW':
        return 'BMW'
    elif cat == 'Volkswagen':
        return 'Volkswagen'
    elif cat == 'Mercedes-Benz':
        return 'Merc'
    else:
        return 'Others'
```

```python
Cars['CarT'] = Cars['car'].map(CarCat)
```

```python
Cars['CarT'].value_counts()
```
